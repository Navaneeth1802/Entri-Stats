```python
#We have to convert the values which are in String to numeric
#Importing the neceassary libraries for transformation
from sklearn.compose import ColumnTransformer
import numpy as np
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler,FunctionTransformer
```

It was observed that the string columns has no natural order or heirarchy so its better to use OHE apart from Cylindernumber and doornumber where there is hierarchy so for those 2 we will be applying OrdinalEncoder

```python
Transformer2=ColumnTransformer(transformers=
                          [('TR1',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),['fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem']),
                           ('TR2',OrdinalEncoder(categories=[['twelve','eight','six','five','four','three','two']]),['cylindernumber']),
                           ('TR3',OrdinalEncoder(categories=[['four','two']]),['doornumber']),
                           ('TR5',StandardScaler(),['wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg'])],
                    remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
#Splitting the model to train and test
x=df.drop('price',axis=1)
y=df['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
#Applying the Transformation and Scaling
x_train_transformed=Transformer2.fit_transform(x_train)
x_test_transformed=Transformer2.transform(x_test)
```

```python
##Importing the ML Models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
#Creating a Dictonary of models so that its easy to Fit and predict
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'Support Vector': SVR()
}
```

```python
#importing the necessary validation_scores
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#Fitting the each model and its result will be stored in as a List.
#This will provide us a better look as all these features will be like a df
results = []

for name, model in models.items():
    model.fit(x_train_transformed,y_train)
    y_pred = model.predict(x_test_transformed)

    results.append({
        'model': name,
        'mse': mean_squared_error(y_test, y_pred),
        'mae': mean_absolute_error(y_test, y_pred),
        'r2_score': r2_score(y_test, y_pred)
    })
results
```

Based on the Results The Random Forest & Gradeint Boosting performed well due to its ensemble nature and tries to learn complex patterns instead of treating it as linear ones also as Random Forest is average of many trees so its less prone to Outliers.

```python
#Feature-Importance
import pandas as pd
import matplotlib.pyplot as plt

feature_names = Transformer2.get_feature_names_out()
# Random Forest Feature Importance
rf_model = models['Random Forest']
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

```

```python
#Gradient_Bossting Feature Importance
rf_model = models['Gradient Boosting']
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

feature_importance_df
```

```python
#Hyper-parameter tuning for GCV
from sklearn.model_selection import GridSearchCV

param={
    'n_estimators':[100,200],
    'max_depth':[10,20],
    'min_samples_split':[1,4]

}
random_f=RandomForestRegressor()
GCV=GridSearchCV(estimator=random_f,param_grid=param,cv=5,scoring='r2')
```

```python
#Fitting the training on GCV
GCV.fit(x_train_transformed,y_train)
```

```python
print(GCV.best_score_)
#SCORE HAS Increased compared to the first 89%
```

```python
#GCV FOR GRADEINT BOOSTING

param_gb={
    'n_estimators':[100,200],
    'learning_rate':[0.01,0.1],
    'max_depth':[3,5]
}

GB=GradientBoostingRegressor()
Gcv_GB=GridSearchCV(estimator=GB,param_grid=param_gb,cv=5,scoring='r2',n_jobs=2)
```

```python
Gcv_GB.fit(x_train_transformed,y_train)
```
