```python
#Plotting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Years=[2019,2020,2021,2022,2023,2024,2025]
City_A=[500000, 550000, 600000, 650000, 700000, 750000, 800000]
City_B=[800000, 850000, 900000, 950000, 1000000, 1050000, 1100000]
City_C=[1000000, 1050000, 1100000, 1150000, 1200000, 1250000, 1300000]
City_D=[1200000, 1250000, 1300000, 1350000, 1400000, 1450000, 1500000]
```

```python
plt.figure(figsize=(12,6))
plt.plot(Years,City_A,marker='o')
plt.plot(Years,City_B,marker='s')
plt.plot(Years,City_C,marker='d')
plt.plot(Years,City_D,marker='^')
plt.xlabel('YEARS')
plt.ylabel('Population')
plt.title("POPULATION OVER YEARS")
```

```python
#BAR CHART
Month=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
Sales=[11860, 10480, 4997, 5523, 13965, 6011, 13158, 9533, 5158, 9058, 11346, 6675]
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange',
              'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray',
              'tab:olive', 'tab:cyan', 'tab:blue', 'tab:red']

plt.bar(Month,Sales,color=bar_colors)
plt.xlabel("Month")
plt.ylabel("SALES")
plt.tight_layout()
plt.show()
```
